```python
#Dependencies
import pandas as pd
import numpy as np
```

```python
#Data source
csv_path = 'purchase_data.csv'
```

```python
#To Pandas data frame
Purchase_df = pd.read_csv(csv_path)
```

```python
#Total players count
total_players = len(Purchase_df["SN"].value_counts())
```

```python
#Data frame to show players count
total_number = pd.DataFrame({"Total Players":[total_players]})
total_number
```

```python
#To get the number of unique items, average price, purchase count, and revenue for purchasing analysis
unique_items = len((Purchase_df["Item ID"]).unique())
average_price = (Purchase_df["Price"]).mean()
total_purchases = (Purchase_df["Purchase ID"]).count()
total_revenue = (Purchase_df["Price"]).sum()
```

```python
#Data frame to show purchasing analysis
purchasing_analysis_df = pd.DataFrame({"Unique Items":[unique_items],
                           "Average Price":[average_price], 
                           "Total Number of Purchases": [total_purchases], 
                           "Total Revenue": [total_revenue]})
purchasing_analysis_df.style.format({'Average Price':"${:,.2f}",
                         'Total Revenue': '${:,.2f}'})
```

```python
#Calculate gender demographics - percentages & counts of male, female & non-disclosed players
gender_groups = Purchase_df.groupby("Gender")
count_gender = gender_groups.nunique()["SN"]
players_percentage = count_gender / total_players * 100
```

```python
#Data frame for gender demographics
gender_demographics = pd.DataFrame({"Percentage of Players": players_percentage, "Total Count": count_gender})
gender_demographics = pd.DataFrame({"Percentage of Players": players_percentage, "Total Count": count_gender})
gender_demographics.index.name = None

#Formatting
gender_demographics.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```

```python
#Calculate purchase count, average purchase price, total purchase value, average purchase total per person by gender
purchase_count = gender_groups["Purchase ID"].count()
avg_purchase_price = gender_groups["Price"].mean()
avg_purchase_total = gender_groups["Price"].sum()
avg_purchase_per_person = avg_purchase_total/count_gender

```

```python
#Data frame for purchase analysis by gender
gender_demographics = pd.DataFrame({"Purchase Count": purchase_count, 
                                    "Average Purchase Price": avg_purchase_price,
                                    "Average Purchase Value":avg_purchase_total,
                                    "Average Purchase Total per Person": avg_purchase_per_person}) 

#Formatting
gender_demographics.index.name = "Gender"
gender_demographics.style.format({"Average Purchase Value":"${:,.2f}",
                                  "Average Purchase Price":"${:,.2f}",
                                  "Avg Purchase Total per Person":"${:,.2f}"})
```

```python
#Binning age demographics
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Assigning bins
Purchase_df["Age Group"] = pd.cut(Purchase_df["Age"],age_bins, labels=group_names)
Purchase_df

```

```python
#Adding new category "Age Group" and calculations for percentage of players and total count
age_grouped = Purchase_df.groupby("Age Group")
total_count_age = age_grouped["SN"].nunique()
percentage_by_age = (total_count_age/total_players) * 100


```

```python
#Data frame for binned age demographics
age_demographics = pd.DataFrame({"Percentage of Players": percentage_by_age, "Total Count": total_count_age})

#Formatting
age_demographics.index.name = None
age_demographics.style.format({"Percentage of Players":"{:,.2f}"})
```

```python
#Calculations for purchasing analysis by age group
purchase_by_age = age_grouped["Purchase ID"].count()
avg_purchase_age = age_grouped["Price"].mean()
total_purchases = age_grouped["Price"].sum()
avg_purchase_person_age = total_purchases/total_count_age



```

```python
#Data frame for purchasing analysis by age
age_demographics = pd.DataFrame({"Purchase Count": purchase_by_age,
                                 "Average Purchase Price": avg_purchase_age,
                                 "Total Purchase Value":total_purchases,
                                 "Average Purchase Total per Person": avg_purchase_person_age})

# Formatting
age_demographics.index.name = None
age_demographics.style.format({"Average Purchase Price":"${:,.2f}",
                               "Total Purchase Value":"${:,.2f}",
                               "Average Purchase Total per Person":"${:,.2f}"})
```

```python
# Grouping purchases by players and running basic calculations
top_spenders = Purchase_df.groupby("SN")
spender_purchase_count = top_spenders["Purchase ID"].count()
avg_purchase_price_spender = top_spenders["Price"].mean()
spender_total_purchase = top_spenders["Price"].sum()

```

```python
#Data frame for top spenders
top_spender_stats = pd.DataFrame({"Purchase Count": spender_purchase_count,
                             "Average Purchase Price": avg_purchase_price_spender,
                             "Total Purchase Value":spender_total_purchase})

#Formatting
formatted_spenders = top_spender_stats.sort_values(["Total Purchase Value"], ascending=False).head()
formatted_spenders.style.format({"Average Purchase Total":"${:,.2f}",
                                 "Average Purchase Price":"${:,.2f}", 
                                 "Total Purchase Value":"${:,.2f}"})
```

```python
#Data frame to include categories for calculating stats for most popular items
items = Purchase_df[["Item ID", "Item Name", "Price"]]
item_stats = items.groupby(["Item ID","Item Name"])
purchase_count_item = item_stats["Price"].count()
purchase_value = (item_stats["Price"].sum()) 
item_price = purchase_value/purchase_count_item

```

```python
#Data frame for most popular items 
most_popular_items = pd.DataFrame({"Purchase Count": purchase_count_item, 
                                   "Item Price": item_price,
                                   "Total Purchase Value":purchase_value})

#Ascending style of data presentation
popular_formatted = most_popular_items.sort_values(["Purchase Count"], ascending=False).head()

#Formatting
popular_formatted.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```

```python
#Data frame for most profitable items 
popular_formatted = most_popular_items.sort_values(["Total Purchase Value"],
                                                   ascending=False).head()
#Formatting
popular_formatted.style.format({"Item Price":"${:,.2f}",
                                "Total Purchase Value":"${:,.2f}"})
```
